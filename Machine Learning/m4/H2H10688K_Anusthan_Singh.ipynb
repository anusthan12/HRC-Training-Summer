### Defining a function to imputate outlier with any value

```python
def check(x,value,up_lt,lo_lt):
    if x>up_lt or x<lo_lt:
        return value
    return x
```

```python
def outlier_imputation(dataset,column_name,value):
    Q1=dataset[column_name].quantile(0.25)
    Q3=dataset[column_name].quantile(0.75)
    IQR=Q3-Q1
    upper_limit=Q3+1.5*IQR
    lower_limit=Q1-1.5*IQR
    return dataset.apply(lambda row:check(row[column_name],value,upper_limit,lower_limit),axis=1)
```

### Using Median to imputate Outlier

```python
median = data['amount_in_usd'].median()
data['amount_in_usd_median']=outlier_imputation(data,'amount_in_usd',median)
```

### Creating a function to convert Delivery day to deliver type (1->fastest,...,7->slowest)

```python
def f1(x):
    if x>=0 and x<5:
        return 1
    elif x>=5 and x<10:
        return 2
    elif x>=10 and x<15:
        return 3
    elif x>=15 and x<20:
        return 4
    elif x>=20 and x<25:
        return 5
    elif x>=25 and x<30:
        return 6
    else:
        return 7
```

### Datetime Manupulation

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    return melt
```

```python
data_temp=difference_in_days(data.drop(['PURCHASE_ORDER_TYPE'],axis=1),10,1,'CUSTOMER_NUMBER','ORDER_CREATION_DATE','amount_in_usd_median')
for i in range(1,11):
    data_temp['Last-'+str(i)+'day_Sales']=data_temp['Last-'+str(i)+'day_Sales'].fillna(0)
```
